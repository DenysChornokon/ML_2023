
Cтворимо власний набір даних для регресії.

```python
import numpy as np

# Створення синтетичного набору даних для регресії
np.random.seed(0)
X = np.random.rand(100, 1) * 10
y = 2 * X + 1 + np.random.randn(100, 1) * 2  # Лінійна залежність з шумом
```

Побудуємо нейронну мережу для регресії та навчимо її на цьому наборі даних.

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# Створення нейронної мережі
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(1,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(1, activation='linear'))

# Компіляція моделі
optimizer = Adam(learning_rate=0.001)  # Зміна швидкості навчання
model.compile(optimizer=optimizer, loss='mean_squared_error')

# Навчання моделі
history = model.fit(X, y, epochs=200, verbose=0, validation_split=0.2)

# Відображення результатів навчання
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
