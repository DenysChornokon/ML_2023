```python
# Словники

data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'San Francisco', 'Los Angeles']
}

df_dict = pd.DataFrame(data_dict)

# Списки

data_list = [
    ['Alice', 25, 'New York'],
    ['Bob', 30, 'San Francisco'],
    ['Charlie', 35, 'Los Angeles']
]

df_list = pd.DataFrame(data_list, columns=['Name', 'Age', 'City'])

# Операція head: Виведення перших рядків

print("DataFrame from Dictionary:")
print(df_dict.head())

print("\nDataFrame from List:")
print(df_list.head())

# Операція describe: Статистичний опис даних

print("\nDescription of DataFrame from Dictionary:")
print(df_dict.describe())

print("\nDescription of DataFrame from List:")
print(df_list.describe())

# Операція iloc: Вибірка даних за індексами

# Вибірка першого рядка
print("\nFirst row using iloc:")
print(df_dict.iloc[0])

# Вибірка конкретного елемента
print("\nElement at row 1, column 'Name' using iloc:")
print(df_dict.iloc[1]['Name'])

# Операція loc: Вибірка даних за мітками

# Вибірка за ім'ям
print("\nSelecting by label using loc:")
print(df_dict.loc[df_dict['Name'] == 'Bob'])
```

```python
# Генерація даних
np.random.seed(42)
X = np.linspace(0, 10, 100)
Y_true = 2 * X + 1 + np.random.normal(scale=2, size=len(X))

# Побудова графіка
plt.scatter(X, Y_true, label='Дані з помилками')
plt.plot(X, 2 * X + 1, color='red', label='Лінійна функція без помилок')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()

# Обчислення прогнозованих значень
Y_pred = 2 * X + 1

# Обчислення метрик помилок
mae = np.mean(np.abs(Y_true - Y_pred))
mse = np.mean((Y_true - Y_pred)**2)

print(f"MAE: {mae}")
print(f"MSE: {mse}")

# Запис результатів у CSV файл
df = pd.DataFrame({'X': X, 'Y_true': Y_true, 'Y_pred': Y_pred, 'MAE': mae, 'MSE': mse})
df.to_csv('results.csv', index=False)
```
