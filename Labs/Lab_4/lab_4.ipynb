```python
# Імпорт необхідних бібліотек
import tensorflow as tf
import numpy as np

# Загальні операції з TensorFlow

# a. Створення тензорів

# Створення скаляра (0-розмірного тензору)
scalar_tensor = tf.constant(5)
print("Скаляр (0D тензор):", scalar_tensor)

# Створення вектора (1D тензору) за допомогою numpy
vector_tensor = tf.constant(np.array([1, 2, 3]))
print("Вектор (1D тензор):", vector_tensor)

# Створення матриці (2D тензору) з випадковими значеннями
matrix_tensor = tf.random.normal(shape=(2, 3))
print("Матриця (2D тензор):", matrix_tensor)

# b. Операції з тензорами

# Операції додавання, віднімання, множення та ділення
tensor_a = tf.constant([[1, 2], [3, 4]])
tensor_b = tf.constant([[5, 6], [7, 8]])

print()
addition_result = tf.add(tensor_a, tensor_b)
print("Додавання:\n", addition_result)

print()
subtraction_result = tf.subtract(tensor_a, tensor_b)
print("Віднімання:\n", subtraction_result)

print()
multiplication_result = tf.multiply(tensor_a, tensor_b)
print("Множення:\n", multiplication_result)

print()
division_result = tf.divide(tensor_a, tensor_b)
print("Ділення:\n", division_result)

# Застосування математичних функцій (синус, косинус)
tensor_c = tf.constant([[0.0, np.pi/2], [np.pi/3, 2*np.pi/3]])

print()
sin_result = tf.math.sin(tensor_c)
print("Синус:\n", sin_result)

print()
cos_result = tf.math.cos(tensor_c)
print("Косинус:\n", cos_result)

# c. Індексація та зрізи

print()
# Звернення до окремих елементів
print("Елемент на позиції (1, 1):", tensor_a[1, 1].numpy())

print()
# Зріз для отримання підмножини даних
subset_tensor = tensor_a[:, 1]
print("Зріз для отримання стовпця з позначкою 1:\n", subset_tensor.numpy())

print()
# d. Зміна форми тензорів (решейп)
reshaped_tensor = tf.reshape(tensor_a, shape=(1, 4))
print("Змінений розмір та форма тензору:\n", reshaped_tensor.numpy())
```

Регресія в TensorFlow

```python
import matplotlib.pyplot as plt

# Згенерувати синтетичний набір даних для регресії
features = np.random.rand(100, 1)  # одна ознака
labels = 2 * features + 1 + 0.1 * \
    np.random.randn(100, 1)  # лінійна залежність з шумом

# Створити модель
model_regression = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,), activation='linear')
])

# Вибрати функцію втрати та оптимізатор
model_regression.compile(optimizer='sgd', loss='mean_squared_error')

# Навчити модель
model_regression.fit(features, labels, epochs=100, verbose=0)

# Оцінити регресійну точність
predictions = model_regression.predict(features)
plt.scatter(features, labels, label='Дані')
plt.plot(features, predictions, color='red',
         linewidth=3, label='Предсказані значення')
plt.xlabel('Ознака')
plt.ylabel('Цільова змінна')
plt.legend()
plt.show()
```

Класифікація в TensorFlow

```python
# Згенерувати синтетичний набір даних для бінарної класифікації
features_classification = np.random.rand(100, 2)  # дві ознаки
labels_classification = (
    features_classification[:, 0] + features_classification[:, 1] > 1).astype(int)

# Створити модель
model_classification = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(2,), activation='sigmoid')
])

# Вибрати функцію втрати, оптимізатор та метрику
model_classification.compile(
    optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Навчити модель
model_classification.fit(features_classification,
                         labels_classification, epochs=100, verbose=0)

# Оцінити класифікаційну точність
accuracy = model_classification.evaluate(
    features_classification, labels_classification, verbose=0)[1]
print("Класифікаційна точність:", accuracy)
```

Візуалізація результів

```python
# Для задачі регресії
import seaborn as sns
from sklearn.metrics import confusion_matrix
plt.scatter(features, labels, label='Дані')
plt.plot(features, predictions, color='red',
         linewidth=3, label='Предсказані значення')
plt.xlabel('Ознака')
plt.ylabel('Цільова змінна')
plt.legend()
plt.show()

# Для задачі класифікації можна використовувати матрицю плутанини

predictions_classification = model_classification.predict(
    features_classification)
conf_matrix = confusion_matrix(
    labels_classification, predictions_classification > 0.5)

sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Предсказані класи')
plt.ylabel('Справжні класи')
plt.show()
```
